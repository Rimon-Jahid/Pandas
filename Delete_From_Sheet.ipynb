Now we will learn how we can delete using (del) function

```python
import pandas as pd
dataframe = pd.read_csv("Screen Time Data.csv")
dataframe.head()
```

```python
dataframe.pop("Date")
dataframe.head()
```

Now we will learn how we can del using drop

```python
import pandas as pd
dataframe = pd.read_csv("Screen Time Data.csv")
dataframe.head()
```

```python
dataframe.drop("Total Screen Time ",axis=1,inplace=True)
dataframe.head()
```

```python
dataframe.drop(columns=["Date","Yoga"],inplace=True)
dataframe.head()
```
